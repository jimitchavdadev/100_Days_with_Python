```python
# rock, paper and scissor code

import os
import random

def clear_console():
    os.system('cls' if os.name == 'nt' else 'clear')

user_score = 0
comp_score = 0

print("Press 'q' followed by Enter to exit")

while True:
    user_input = input("Type your choice (0 for Rock, 1 for Paper, 2 for Scissors) or 'q' to quit: ")
    if user_input == 'q':  # Check if the input is 'q'
        print("Exiting...")
        break  # Exit the loop

    if user_input not in ['0', '1', '2']:
        print("Invalid input, please enter 0, 1, 2, or 'q'.")
        continue
    
    number = int(user_input)
    comp_choice = random.randint(0, 2)

    choices = ["Rock", "Paper", "Scissors"]

    print(f"You chose: {choices[number]}")
    print(f"Computer chose: {choices[comp_choice]}")

    if number == comp_choice:
        print("It's a tie!")
    elif (number == 0 and comp_choice == 2) or (number == 1 and comp_choice == 0) or (number == 2 and comp_choice == 1):
        user_score += 1
        print("You win this round!")
    else:
        comp_score += 1
        print("Computer wins this round!")

    print(f"Your score: {user_score}")
    print(f"Computer's score: {comp_score}")

    clear_console()
```
